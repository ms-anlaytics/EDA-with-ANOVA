### Import Data

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
#read data into dataframe
churn_df = pd.read_csv('churn_raw_data.csv',dtype={'locationid':np.int64})

```

Data is imported into the project. 

## Clean Data

```python
#rename survey data
churn_df.rename(columns = {'item1':'Timely_Response','item2':'Timely_Fixes','item3':'Timely_Replacements','item4':'Reliability','item5':'Options','item6':'Respectful_Response','item7':'Courteous_Exchange','item8':'Active_Listening'}, inplace = True)
churn_df.head(1)
```

Columns of potential interest are renamed with meaningful names. 

```python
#reduce dataset to relevant numeric data 
dropCols = ['Unnamed: 0', 'CaseOrder', 'Lat', 'Lng', 'Children', 'Age', 'Outage_sec_perweek', 'Email', 'Contacts','Yearly_equip_failure', 'Active_Listening']
churn_df = churn_df.drop(axis=1, labels=dropCols)
```

Unnecessary data is dropped from the dataset and stored in <code>churn_df</code>

```python
# impute missing data with mean
numeric_data=churn_df.columns[churn_df.dtypes != 'object']
churn_df[numeric_data] = churn_df[numeric_data].fillna(churn_df.mean())
churn_df[numeric_data].head(1)
```

The numerical data is identified and any missing fields are imputed with the mean value. 

```python
#define categoric data

category_data = churn_df.columns[churn_df.dtypes == 'object']

# Fill missing data with "missing data"
churn_df[category_data] = churn_df[category_data].fillna('Missing Data')
churn_df[category_data].head(1)
```

The categorical data is now identified and any Null values are replaced with <code>“Missing Data”</code>

## Data Analysis: ANOVA Test

```python
import random
# filter the customers who discontinued service in the last month (De-Yu, 2021)
discontinued_df = churn_df[churn_df['Churn'] == 'Yes']
discontinued_df.head(1)
```

The dataset is filtered where the<code>Churn</code>column contains the variable <code>Yes</code>. This is to guarantee that the sample data for the ANOVA test contains only the population of customers who cancelled service. 

```python
# random sample for 2000 customers (De-Yu, 2021)
customer_id = list(discontinued_df['Customer_id'].unique())
random.seed(30) # set a seed so that everytime we will extract same sample
sample_customer_id = random.sample(customer_id , 2000)
sample_df = discontinued_df[discontinued_df['Customer_id'].isin(sample_customer_id)].reset_index(drop=True)
sample_df.head(1)
```

Now a random sample of 2000 unique customers is taken from discontinued customers and stored into the data frame <code>sample_df</code>.

```python
# find two variables of interest (De-Yu, 2021)
sample_df1 = sample_df[['State', 'Income']]
groups = sample_df1.groupby('State').count().reset_index()

groups
```

Two variables “<code>State</code>” and “<code>Income</code>” are stored in the “<code>groups</code>” data frame.

```python
#Normality Assumption Check (De-Yu, 2021)

import scipy.stats as stats
import matplotlib.pyplot as plt

unique_State = sample_df1['State'].unique()
for State in unique_State:
    stats.probplot(sample_df1[sample_df1['State'] == State]['Income'], dist="norm", plot=plt)
    plt.title("Probability Plot - " +  State)
    plt.show()
```

Normality Assumption Check is performed on unique states contain in the data. (De-Yu, 2021)

```python
#Homogeneity of variance Assumption Check (De-Yu, 2021)

# calculate ratio of the largest to the smallest sample standard deviation
ratio = sample_df1.groupby('State').std().max() / sample_df1.groupby('State').std().min()
ratio
print('The ratio of the largest to the smallest sample standard deviation is '+str(ratio.Income))
```

```python
#ANOVA backbone table
data = [['Between Groups', '', '', '', '', '', ''], ['Within Groups', '', '', '', '', '', ''], ['Total', '', '', '', '', '', '']] 
anova_table = pd.DataFrame(data, columns = ['Source of Variation', 'Sum of Squares', 'Degrees of Freedom', 'Mean Squares', 'F-value', 'P-value', 'F critical']) 
anova_table.set_index('Source of Variation', inplace = True)

# calculate SSTR (sum of the squares due to treatment) and update anova table
x_bar = sample_df1['Income'].mean()
SSTR = sample_df1.groupby('State').count() * (sample_df1.groupby('State').mean() - x_bar)**2
anova_table['Sum of Squares']['Between Groups'] = SSTR['Income'].sum()

# calculate SSE (sum of squares due to error) and update anova table
SSE = (sample_df1.groupby('State').count() - 1) * sample_df1.groupby('State').std()**2
anova_table['Sum of Squares']['Within Groups'] = SSE['Income'].sum()

# calculate SSTR and update anova table
SSTR = SSTR['Income'].sum() + SSE['Income'].sum()
anova_table['Sum of Squares']['Total'] = SSTR

# update degree of freedom
anova_table['Degrees of Freedom']['Between Groups'] = sample_df1['State'].nunique() - 1
anova_table['Degrees of Freedom']['Within Groups'] = sample_df1.shape[0] - sample_df1['State'].nunique()
anova_table['Degrees of Freedom']['Total'] = sample_df1.shape[0] - 1

# calculate Mean Squares
anova_table['Mean Squares'] = anova_table['Sum of Squares'] / anova_table['Degrees of Freedom']

# calculate F 
F = anova_table['Mean Squares']['Between Groups'] / anova_table['Mean Squares']['Within Groups']
anova_table['F-value']['Between Groups'] = F

# p-value
anova_table['P-value']['Between Groups'] = 1 - stats.f.cdf(F, anova_table['Degrees of Freedom']['Between Groups'], anova_table['Degrees of Freedom']['Within Groups'])

# F critical 
alpha = 0.05
# possible types "right-tailed, left-tailed, two-tailed"
tail_hypothesis_type = "two-tailed"
if tail_hypothesis_type == "two-tailed":
    alpha /= 2
anova_table['F critical']['Between Groups'] = stats.f.ppf(1-alpha, anova_table['Degrees of Freedom']['Between Groups'], anova_table['Degrees of Freedom']['Within Groups'])

# Final ANOVA Table 
anova_table

#(De-Yu, 2021)
```

The results of the ANOVA test have been printed in the table above. An important variable to pay attention to is <code>P-value</code> as it will be used in the hypothesis test below to determine if the regionality of the customer base has an impact on the population who cancel service with the organization. (De-Yu, 2021)

```python
# Hypothosis Testing Result (De-Yu, 2021)

conclusion = "Failed to reject the null hypothesis."
if anova_table['P-value']['Between Groups'] <= alpha:
    conclusion = "Null Hypothesis is rejected."
    
conclusion
```

The results of the hypothesis test printed in the variable <code>conclusion</code> are now printed in the condition of the test result. In this case the Null Hypothesis was maintained, meaning that the state a customer resides in has no association to whether they continue service with the cellular company.

## Bivariate Statistics

### Distribution of continuous variables.

<p>The Pearson Correlation test will determine a relationship of the bivariate variables of “Income” and “Monthly Charges”. The resulting correlation variables will range between -1 and 1, where zero can be interpreted that there is no correlation at all. A negative number means the correlation is inverse and positive number means the correlation is parallel. (Akiode, 2021) </p>

```python
#Using Pearson Method to test if a linear relationship exists between MonthlyCharge and Income
import numpy as np
from scipy import stats
from scipy.stats import pearsonr
billingY=sample_df['MonthlyCharge']
incomeX=sample_df['Income']
stats.pearsonr(incomeX, billingY)
```


<p>The result of the test, <code>stats.pearsonr(billing, income)</code> yields the Pearson correlation statistic<code>-0.03171465699347514</code> and the P-value of the data <code>0.15625135385234043</code>. The conclusion, the variables correlate negatively and increase and decrease inversely. (Akiode, 2021) </p>


```python
#Bivariate Visual Distribution between MonthlyCharge and Income
fig =sns.displot(data=sample_df, x="Income", y="MonthlyCharge", cbar=True, height=10,aspect=1, color="red")
plt.xlabel("Income")
plt.ylabel("MonthlyCharge")
plt.title("Income vs Monthly Bill", fontweight ='bold')
plt.show(fig)
```

```python
#Using Pearson Method to test if a linear relationship exists between MonthlyCharge and Bandwidth_GB_Year

VarY=sample_df['MonthlyCharge']
VarX=sample_df['Bandwidth_GB_Year']
stats.pearsonr(VarX, VarY)
```

The Pearson correlation statistic<code>0.2989893843157122</code> and the P-value of the data <code>1.395128319919986e-42</code>.
<p/>The conclusion is the variables do positively correlate and increase and decrease in parallel. (Akiode, 2021)

```python
#Bivariate Visual Distribution between MonthlyCharge and Bandwidth_GB_Year

fig =sns.displot(data=sample_df, x="Bandwidth_GB_Year", y="MonthlyCharge",cbar=True, height=10,aspect=1, color="red")
plt.xlabel("Bandwidth_GB_Year")
plt.ylabel("MonthlyCharge")
plt.title("Bandwidth Usage vs Monthly Bill", fontweight ='bold')
plt.show(fig)
```

### Distribution of categorical variables

In the Bivariate distribution analysis of categorical variables, the data is first arranged into a contingency table to essentially count the instances of associated variable crosses. Then the contingency data is checked for relationships with a Pearson correlation method from Python using, <code>corr()</code>. In observing the resulting table, a result of 1 means that the variable is correlating with itself. As before a negative or positive result is either an inverse or parallel correlation between the variables. Zero still means no correlation at all. (Ranjan, 2021)

```python
# contingency table to reveal a relationship between Education and State
Result=pd.crosstab(index=sample_df['Education'],columns=sample_df['State'])
Result
```

The contingency table counts the pair crosses of each categorical variable.

```python
# Categorical Correlation Table Using Pearson Method to show distribution of education level by state

Result.corr(method ='pearson')
```

The Pearson correlation table now shows the correlation coefficient between <code>State</code> and <code> Education</code>. This indicates that customer education level shares a relationship between states. A state education level correlates with itself, indicated by a result of 1. 

```python
#Clustermap of contingency table Result
fig=sns.clustermap(Result)
plt.xlabel("Education")
plt.ylabel("Count")
plt.title("Education Level", fontweight ='bold')
plt.show(fig)
```

A cluster map of the resulting correlation table is displayed to show the frequency of correlation between states and levels of education for customers leaving the services of the company.

```python
# contingency table to reveal a relationship between Job and Area
Result2=pd.crosstab(index=sample_df['Area'],columns=sample_df['Job'])
Result2
```

The contingency table now counts the associations of <code>Area</code> and <code>Job</code> and displays the information in the table. 

```python
# Categorical Correlation Table Using Pearson Method to show distribution of Job by Area
Result2.corr(method ='pearson')
```

The Pearson correlation table now displays relationships in the distribution of data as it relates to the job a customer reports as an occupation. The correlation of occupation as it relates to the area, customers report living in, is then displayed in the table. Again, zero is no correlation, negative and positive numbers correlate inversely or positively, and 1 indicates the categorical correlation of a variable to is self.  

```python
#Clustermap of cPearson correlation table Result2
fig=sns.clustermap(Result2)
plt.xlabel("Job")
plt.ylabel("Count")
plt.title("Occupation", fontweight ='bold')
plt.show(fig)
```

Again, a cluster map displays the Pearson relationship coefficient by shade of the color red for customer occupation and area.
